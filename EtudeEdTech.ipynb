```python
# cette fonction donne un pourcentage de valeur manquantes sur l'ensemble du dataframe
def taux_na(dataframe,value=False):
    values = dataframe.isna().values.ravel()
    taux = round((values.sum() / len(values))*100,2)
    if value == False : # par défaut sous la forme d'une phrase
        return(f'Il y a environ {taux} % de valeurs manquantes dans le dataframe.')
    elif value == True: # mais peut renvoyer seulement le pourcentage
        return taux

# cette fonction fournit plus de détails en donnant le pourcentage de valeurs manquantes par colonne
def taux_na_colonnes(dataframe):
    colsum = dataframe.isna().sum()
    colonnes = colsum.index
    valeurs = colsum.values
    pourcentage = np.round(((valeurs / dataframe.shape[0])*100),2)
    data_item = {'colonnes':colonnes,'taux NAN':valeurs,'pourcentage de NAN':pourcentage}
    with pd.option_context('display.max_rows',dataframe.shape[1]):
        display(pd.DataFrame(data_item).sort_values(by='pourcentage de NAN',ascending=False))
# Cette fonction permet de visualiser un tableau dans son ensemble avec les zones contenant des données         
def plot_na(dataframe,size=(20,10)):
    plt.figure(figsize=size)
    sns.heatmap(dataframe.isna(),cbar=False);
```

## Calcul du rang bivarié 
le rang bivarié est une formule qui nous permet de classer des éléments sur la base de plusieurs indicateurs.

Pour ce faire nous employons cette formule: 

Qi = 3/4 + ∑j u(Xi−Xj)u(Yi−Yj)


# Calcul du rang bivarié 
le rang bivarié est une formule qui nous permet de classer des éléments sur la base de plusieurs indicateurs.

(source : https://blogs.sas.com/content/iml/2021/04/21/bivariate-ranks.html)

La formule nous permet de calculer le nombre de fois qu'un élément s'est retrouvé au dessus ou en dessous d'autres éléments 

```python
# x négatif : Xi est en dessous de Xj
# x = 0 : ex aequo
# x positif : Xi supérieur à Xj
def u(x):
    mask_1 = x > 0
    mask_05 = x == 0
    mask_0 = x < 0
    x[mask_1] = 1
    x[mask_05] = 0.5
    x[mask_0] = 0 
    return x
```

```python
# implémentation en code de la formule
def bivariate_rank(x,y):
    rank = []
    for i in range(len(x)):
         rank.append((0.75 + np.sum(u(x[i]-x) * u(y[i]-y))))
    return rank
```

```python
# Construction d'une table qui classe les pays sur la base du rang bivarié calculé sur les deux indicateurs passés en argument
def get_ranking(tableIndic1,tableIndic2):
    # sauvegarde des noms des indicateurs
    indicCode1 = tableIndic1['Indicator Code'].reset_index(drop=True)[0]
    indicCode2 = tableIndic2['Indicator Code'].reset_index(drop=True)[0]
    # filtrage des tables
    tableIndic1 = tableIndic1.loc[:,['Table Name','Indicator Value']]
    tableIndic2 = tableIndic2.loc[:,['Table Name','Indicator Value']]
    # Construction de la table
    table = pd.merge(tableIndic1,tableIndic2,left_on='Table Name',right_on='Table Name')
    table.rename(columns={"Indicator Value_x":indicCode1,"Indicator Value_y":indicCode2},inplace=True)
    table['rank'] = bivariate_rank(table.iloc[:,1],table.iloc[:,2])
    table.sort_values(by='rank',ascending=False,inplace=True)
    table.reset_index(drop=True,inplace=True)
    return table
```

```python
def plot_cross_indic(table,indice):
    indic1 = table.columns[1]
    indic2 = table.columns[2]
    max_rank = table['rank'].max()
    max_x_pop = table[indic1].max()
    max_y_pop = table[indic2].max()

    fig, ax = plt.subplots(2,1,
                           figsize=(15,13),
                           gridspec_kw={'height_ratios':[1,4]})
    plt.subplots_adjust(hspace=0.3)
    cmap = sns.cubehelix_palette(as_cmap=True)
    ax[0].set_xticks(np.linspace(0,max_rank,8))
    ax[0].set_title(f'Classement des pays (Accès internet - {indice})',fontsize=15)

    sns.barplot(ax=ax[0],
                x=table['rank'][0:9],
                y=table['Table Name'][0:9],
                palette='Purples_r')
    ax[0].set_xlabel('Rang bivarié (plus grand = en tête)')
    ax[0].set_ylabel('Pays')
    ax[1].set_title(f'Accès internet vs {indice}',fontsize=15)
    ax[1].set_ylabel('Accès internet \n (Quantité moyenne sur 100)')
    ax[1].set_xlabel(indice)
    points = ax[1].scatter(x = table[indic1],
                  y = table[indic2],
                  c=table['rank'],
                  cmap=cmap)

    ax[1].set_xticks(np.linspace(0,max_x_pop,5))
    ax[1].set_yticks(np.linspace(0,max_y_pop,8))
    ax[1].xaxis.set_major_formatter(plt.FuncFormatter(readable_number))
    ax[1].yaxis.set_major_formatter(plt.FuncFormatter(readable_number))
   

    cbar = fig.colorbar(points,ticks=[1,table['rank'].max()])
    cbar.ax.set_yticklabels(['Score faible','Score élevé'])

    labels = table['Table Name'][0:5]
    for i in range(0,len(labels)):
            plt.annotate(
                labels[i],
                xy=(table.loc[i,indic1],table.loc[i,indic2]),
                xytext=((i+8)**2.1,0),
                textcoords='offset points', ha='right', va='bottom',
                bbox=dict(boxstyle='round,pad=0.5', fc='purple', alpha=0.2),
                arrowprops=dict(arrowstyle='simple',
                                connectionstyle='arc3,rad=0',
                                alpha=0.5))
```

**Pour le calcul final du score nous utiliserons la population réelle**

Ici nous utilisons la population totale du pays et le pourcentage afin de calculer une estimation de la population réelle ayant accès à internet. 

```python
country_pop = dataset[dataset['Indicator Code']=='SP.POP.TOTL'].loc[:,['Table Name','Indicator Value']]
country_pop.rename(columns={'Indicator Value':'pop_total'},inplace = True)
internet_pop = internet_pop.merge(country_pop)
```

```python
# calcul du pourcentage et injection des nouvelles informations dans la table
internet_pop['Indicator Value'] = internet_pop['Indicator Value']/100 * internet_pop['pop_total']
```

Calcul du rang avec population estimée d'utilisateurs internet

```python
popYoung_internet_pop = get_ranking(popYoung,internet_pop)
popTot_internet_pop = get_ranking(popTot,internet_pop)
popUniv_internet_pop = get_ranking(univ,internet_pop)
popSecondary_internet_pop = get_ranking(secondary,internet_pop)
```

```python
list_rankings = [popYoung_internet_pop,
            popTot_internet_pop,
            popUniv_internet_pop,
            popSecondary_internet_pop]
```

```python
rankings = []
for table in list_rankings:
    crossIndic = table.columns[1]
    table = table.loc[:,['Table Name','rank']]
    table['cross_indic'] = crossIndic
    rankings.append(table)
```

```python
rankings = pd.concat(rankings).reset_index(drop=True)
rankings['coefs'] = 1 
```

Les coefficients reflètent la logique du classement: 
* Des pays avec un bon taux de pénétration internet, la condition sinéquanone à l'implentation de l'entreprise. = 0.5 (moins important mais peut comporter des oportunités)
* Une population importante surtout dans la tranche d'age des lycéens et étudiants = 1 (le plus important)
* Un système scolaire développé ouvrant des opportunités de cours de soutien = 0.25 + 0.25  (Est suceptible de passer par nos servies mais pas au même degré qu'une population éduquée 100% en ligne)

```python
rankings.loc[rankings[rankings['cross_indic']=='SP.POP.TOTL'].index.values,'coef'] = 0.25
rankings.loc[rankings[rankings['cross_indic']=='SE.TER.ENRL'].index.values,'coef'] = 0.25
rankings.loc[rankings[rankings['cross_indic']=='UIS.E.3'].index.values,'coef'] = 0.25
```

```python
rankings['weighted_rank'] = rankings['rank'] * rankings['coef']
```

```python
ranked_countries = (rankings.groupby('Table Name',as_index=False)['weighted_rank'].sum()).sort_values(by='weighted_rank',ascending=False)
```
